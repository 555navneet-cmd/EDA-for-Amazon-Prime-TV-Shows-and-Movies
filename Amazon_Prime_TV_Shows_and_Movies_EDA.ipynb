# **Project Summary -**

The project's objective was to conduct an Exploratory Data Analysis (EDA) on Amazon Prime Video's content library. The goal was to gain a comprehensive understanding of the platform's content landscape, identify key trends, and extract actionable insights to inform content strategy, boost user engagement, and drive business growth.

Two primary datasets were utilized for this analysis: credits.csv (df_1), containing information about the cast and crew associated with titles, and titles.csv (df_2), which provided details about each movie and TV show, including title, type, description, release year, ratings (IMDb and TMDb), genres, and production countries.

**The data wrangling phase involved several crucial steps to ensure data quality and suitability for analysis:**

**1.	Duplicate Value Handling:** Duplicate rows were identified and removed from both df_1 (56 duplicates) and df_2 (3 duplicates) to prevent skewed analysis.

**2.	Missing Value Treatment:** Columns with a high percentage of missing values, specifically 'age_certification' and 'seasons' in df_2, were dropped to avoid introducing noise or bias. Missing values in the 'description' column were imputed with a placeholder, 'No description available'. Numerical rating columns like 'imdb_score', 'imdb_votes', 'tmdb_popularity', and 'tmdb_score' had their missing values filled with their respective medians to maintain data distribution. The 'imdb_id' column's missing entries were replaced with 'Unknown'. Lastly, rows in df_1 where the 'character' column was null were dropped, as this information was critical for actor/director analysis.

**Key visualizations and the insights derived provided a clear picture of Amazon Prime's content:**

**Content Split:** Movies dominate the platform (86.2%) compared to TV shows (13.8%).

**Top Genres:** Drama leads with 4,762 titles, followed by Comedy (2,987) and Thriller (2,119), showing strong audience preference.

**Highest-Rated Titles:** Top movies include Pawankhind and Jai Bhim; top shows include Water Helps the Blood Run and The Chosen.

**Exact Volume:** The library contains 8,511 movies and 1,357 TV shows.

**Release Trend:** Content output grew steadily, peaking in 2021 (856 titles), with a slight decline in 2022.

**Ratings Alignment:** IMDb and TMDb scores show strong positive correlation, indicating consistent quality assessment.

**Prolific Contributors:** Frequent contributors include George 'Gabby' Hayes and Roy Rogers, mainly from older content.

**Top Production Countries:** Movies mainly come from the United States, India, and United Kingdom; shows are led by the US, UK, Japan, and China.

**Overall Growth:** Both movies and shows show an upward release trend, with movies consistently outnumbering TV series.

**Based on these insights, the following business recommendations were formulated:**

**Balance Content Mix:** Expand high-quality TV series while continuing to invest in popular genres (Drama, Comedy, Thriller) and top-rated titles on Amazon Prime Video.

**Boost Discovery & Marketing:** Highlight top-rated content and launch curated collections around prolific actors and directors to improve engagement.

**Track Content Trends:** Maintain steady release momentum, analyze dips (e.g., 2022), and use IMDb–TMDb rating alignment for smarter acquisition and targeted promotion.

**Expand Globally:** Strengthen partnerships in the United States, India, and United Kingdom, while increasing TV show investment in Japan and China.

**Fix Content Imbalance:** Prioritize growing the TV series library to meet episodic-content demand and compete more strongly with rival platforms.

These recommendations aim to refine Amazon Prime Video's content strategy, drive engagement, and ensure sustained business growth by catering to diverse audience preferences and capitalizing on market trends.

### Data Wrangling Code

````python
# Create a copy of df_2
df_2_copy = df_2.copy()

# Drop the 'age_certification' column from the copied DataFrame
df_2_copy.drop(columns=['age_certification'], inplace=True)
````

````python
# Drop the 'seasons' column from the copied DataFrame
df_2_copy.drop(columns=['seasons'], inplace=True)
````

````python
# Fill missing 'description' with a placeholder
df_2_copy['description'].fillna('No description available', inplace=True)
````

````python
# Fill missing numerical columns with their median
for col in ['imdb_score', 'imdb_votes', 'tmdb_popularity', 'tmdb_score']:
    df_2_copy[col].fillna(df_2_copy[col].median(), inplace=True)
````

````python
# Fill missing 'imdb_id' with 'Unknown' in df_2_copy
df_2_copy['imdb_id'].fillna('Unknown', inplace=True)
````

````python
# Wrangling codes:

#Proportion of content type i.e. Movies vs Shows
type_counts = df_2_copy['type'].value_counts()

# Get the top 10 genres
# Function to safely parse the genre string into a list
def parse_genres(genres_str):
    try:
        return ast.literal_eval(genres_str)
    except (ValueError, SyntaxError):
        return []
# Apply the function to the 'genres' column to get a list of lists
df_2_copy['parsed_genres'] = df_2_copy['genres'].apply(parse_genres)
# Explode the list of genres into separate rows and count their occurrences
genre_counts = df_2_copy['parsed_genres'].explode().value_counts()

#Top 5 IMDb scores for Movies
movies_df = df_2_copy[df_2_copy['type'] == 'MOVIE']

#Top 5 IMDb scores for Shows
shows_df = df_2_copy[df_2_copy['type'] == 'SHOW']

#Title Count Per Release Year (Horizontal Bar Chart)
# Calculate title count per release year and sort by count in descending order
title_count_per_year = df_2_copy['release_year'].value_counts().sort_values(ascending=False).head(10)

#Top 10 Production Countries for Movies
movies_df = df_2_copy[df_2_copy['type'] == 'MOVIE'].copy()
def parse_countries(countries_str):
    try:
        return ast.literal_eval(countries_str)
    except (ValueError, SyntaxError):
        return []
movies_df['parsed_countries'] = movies_df['production_countries'].apply(parse_countries)
country_counts = movies_df['parsed_countries'].explode().value_counts()

#Top 10 Production Countries for shows
shows_df = df_2_copy[df_2_copy['type'] == 'SHOW'].copy()
def parse_countries(countries_str):
    try:
        return ast.literal_eval(countries_str)
    except (ValueError, SyntaxError):
        return []
shows_df['parsed_countries'] = shows_df['production_countries'].apply(parse_countries)
country_counts_shows = shows_df['parsed_countries'].explode().value_counts()

#Actor/Director participation by volume
roles_to_filter = ['ACTOR', 'DIRECTOR']
filtered_df_1_copy = df_1_copy[df_1_copy['role'].isin(roles_to_filter)]
top_10_actors_directors = filtered_df_1_copy['name'].value_counts().head(10)
print("Top 10 Actors/Directors by Participation:")
print(top_10_actors_directors)
````

### What all manipulations have you done and insights you found?

**Duplicate Value Handling:**

**Missing Value Treatment:**

*   Columns **age_certification** and **seasons** were dropped due to a high percentage of missing values.

*   Missing values in the **description** column were filled with the placeholder string 'No description available'.

*   Numerical columns such as **imdb_score, imdb_votes, tmdb_popularity, and tmdb_score** had their missing values imputed with their respective **median values**.

*   Missing **imdb_id** values were replaced with 'Unknown'.

*   Rows containing null values in the **character** column were dropped.

**Parsing String Representations of Lists:**

*   The **genres** column in the titles dataset, which contained string representations of lists (e.g., "['comedy', 'drama']"), was parsed using ast.literal_eval to convert them into actual Python list objects, enabling easier manipulation and analysis of individual genres.

*   The **production_countries** column in the titles dataset was also parsed from string representations to actual list objects to facilitate analysis of content origins.
